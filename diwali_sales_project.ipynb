# **Exploratry Data Analysis**

```python
df.columns
```

## **Gender**

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount',data=sales_gen)
```

from above graphs we can see that most of the buyers are female and purchasing power of females are greater than men

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount',data=sales_age)
```

# **from above graphs we can see the buyers are in the age group of 26-35**

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.set(rc ={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sns.set(rc ={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

**from above graphs it is  can be seen most money is spent by utarpradesh,maharastra,karnatka**

```python
ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
  ax.bar_label(bars)
sns.set(rc={'figure.figsize':(3,4)})
```

```python
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc ={'figure.figsize':(5,5)})
sns.barplot(data=sales_ms,x='Marital_Status',y='Amount',hue='Gender')
```

# **from above graph we can say that married Females are shopping more and have more purchasing power as well**

```python
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,7)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_ocup=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.barplot(data=sales_ocup,x='Occupation',y='Amount')
sns.set(rc=({'figure.figsize':(20,3)}))
```

# **we can see from above graphs that most of the buyers are from it sector, healthcare and aviation.**

```python
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
sns.set(rc=({'figure.figsize':(35,5)}))
```

```python
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by=('Amount'),ascending=False).head(10)
sns.barplot(data=sales_pc,x='Product_Category',y='Amount')
sns.set(rc=({'figure.figsize':(20,3)}))
```

**from above graphs we can seethat most orders are of clothing and apparel while most money is spent on food.**

```python
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_pid,x='Product_ID',y='Orders')
sns.set(rc=({'figure.figsize':(15,2)}))
```
